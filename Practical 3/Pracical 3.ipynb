# Flight Price Prediction (EDA + Feature Prediction)

```python
# import basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# this data set has date time 
# here we will mostly do feature engineering
# catch one feature and do the necssary changes
# Drop the null value:- final_df.dropna()
```

```python
# import the training and test data set
train_df=pd.read_excel('Data_Train.xlsx')
test_df=pd.read_excel('Test_set.xlsx')
```

```python
train_df.shape,test_df.shape
```

```python
# Combine train and test data
final_df=pd.concat([train_df, test_df])
```

```python
final_df
```

```python
final_df.info()
```

```python
final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
# From Date_of_Journey, we are going to derive date, month and year
# we will split and doindexing
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head() # Date , month and year column created
```

```python
final_df.info()
```

```python
# convert the date month and year into int
final_df['Date']=final_df['Date'].astype('int64')
final_df['Month']=final_df['Month'].astype('int64')
final_df['Year']=final_df['Year'].astype('int64')
```

```python
final_df.info()
```

```python
# drop date of journey feature
final_df=final_df.drop('Date_of_Journey',axis=1)
```

```python
final_df.head(3)
```

```python
# now focus on arrival time and departure time
# try to catch similar type of fields and do it together like the above one
# in arrival we donnot require the date
```

```python
final_df['Arrival_Time'].str.split(' ')
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(2)
```

```python
final_df.isnull().sum()
```

```python
# From arrival time , we will separate hour and min
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_minute']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
# then convert the hour and minute to int
final_df['Arrival_hour']=final_df['Arrival_hour'].astype('int64')
final_df['Arrival_minute']=final_df['Arrival_minute'].astype('int64')
```

```python
final_df.info()
```

```python
# Drop the arrival time
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
# Let us do for Dep_Time 

# From arrival time , we will separate hour and min
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute']=final_df['Dep_Time'].str.split(':').str[1]

# then convert the hour and minute to int
final_df['Dep_hour']=final_df['Dep_hour'].astype('int64')
final_df['Dep_minute']=final_df['Dep_minute'].astype('int64')

# Drop the arrival time
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(2)
```

```python
final_df.info()
```

```python
# Let's Do the route
final_df.Route.value_counts()
# Instead of doing this, we will be using total stops
```

```python
# Total_stops
final_df['Total_Stops'].unique()
```

```python
# how many null value
final_df['Total_Stops'].isnull().sum()
```

```python
final_df.Total_Stops.value_counts()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, 'nan':2, '4 stops':4})
```

```python
final_df.Total_Stops.value_counts()
```

```python
final_df.head(2)
```

```python
# Drop the route column 
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Additional_Info'].value_counts()
```

```python
# We will work with duration , separated hour and minute

# Let us do for Duration

# From arrival time , we will separate hour and min
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_minute']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]

```

```python
# 5m error( It is an error , so we drop it )
final_df[final_df['Duration_hour']=='5m']
```

```python
# drop the 5m
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
# Drop the null value in minute column
final_df.dropna(subset=['Duration_minute'],inplace=True)
```

```python
# then convert the hour and minute to int
final_df['Duration_hour']=final_df['Duration_hour'].astype('int64')
final_df['Duration_minute']=final_df['Duration_minute'].astype('int64')

```

```python
# Drop the arrival time
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
# Rest all are category features 
# Airline
final_df.Airline.unique()
```

```python
# I will do label encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
final_df['Airline']=label_encoder.fit_transform(final_df['Airline'])
final_df['Source']=label_encoder.fit_transform(final_df['Source'])
final_df['Destination']=label_encoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=label_encoder.fit_transform(final_df['Additional_Info'])
```
